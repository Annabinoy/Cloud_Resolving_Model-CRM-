```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
dimen=pd.read_csv('test.csv',header=None,nrows=1)
data=pd.read_csv('test.csv', header=None, skiprows=1)
x=[]
y=[]
z=[]
for i in range(int(dimen[1])):
    for j in range(int(dimen[2])):
        for k in range(int(dimen[3])):
            x.append(i)
            y.append(j)
            z.append(k)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
def plotter(t=0):
    fig=plt.figure(figsize=(20,14))
    ax=fig.add_subplot(111,projection='3d')
    c=data[t]
    ax.set_xlabel('Spatial co-ordinate X')
    ax.set_ylabel('Spatial co-ordinate Y')
    ax.set_zlabel('Spatial co-ordinate Z')
    surf=ax.scatter(x,y,z,c=c,cmap=plt.hot(), vmin=0, vmax=1.5)
    t=fig.colorbar(surf,pad=0.20,shrink=0.85)
    t.set_label('Temperature')
    
    plt.savefig('fig10')
    plt.show()
plotter(39)
```

```python
import ipywidgets as widgets
widgets.interact(plotter, t=widgets.Play(min=0,max=39))
```
